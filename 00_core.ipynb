```python
#export
def say_hello(to):
    "`say_hello` to {to}...with a couple of exceptions."
    if to == 'Jimmy McGill':
        return 'S\'all good, man.'
    elif to == 'Jesse Pinkman':
        return 'Yo'
    else:
        return f'Hello {to}!'
```

An example:

```python
say_hello('Everybody!')
```

And a test:

```python
assert say_hello('Everybody!') == 'Hello Everybody!!'
```

```python
assert say_hello('Jimmy McGill') == 'S\'all good, man.'
```

```python
assert say_hello('Jesse Pinkman') == 'Yo'
```

```python
#export
class HelloSayer:
    "Say hello to `to` using `say_hello`"
    def __init__(self, to): self.to = to

    def say(self):
        "Do the saying with `say`"
        return say_hello(self.to)
```

Example:

```python
greet = HelloSayer('everybody')
greet.say()
```

```python
HelloSayer('Jimmy McGill').say()
```

```python
HelloSayer('Jesse Pinkman').say()
```
